&nbsp;
#### GENERAL TOTAL

```python
# get question item total, question stem, question item, and stem/item ids 
totals = data[['Question Stem Id', 'Question Item Id', 'Question Stem', 'Question Item', 'Question Response', 'Count']]
totals = totals.loc[data['Demographic Category'] == 'Undergrad Grad'].drop_duplicates(ignore_index = True)
```

```python
totals = totals[totals['Question Item Id'].str.contains(question_stem, case=False)]

# for rows with the same item_id and question response, consolidate into one row by adding the counts 
total = totals.groupby(['Question Item Id', 'Question Response', ]).sum('Count')
total.head(2)
```

&nbsp;
#### UG COUNTS

```python
ugtotals = data[['Question Stem Id', 'Question Item Id', 'Question Stem', 'Question Item', 'Question Response', 'Count', 'Undergrad Grad']]
ugtotals = ugtotals.loc[data['Demographic Category'] == 'Undergrad Grad'].drop_duplicates(ignore_index = True)
```

```python
ugtotals = ugtotals[ugtotals['Question Item Id'].str.contains(question_stem, case=False)]

ugtotal = ugtotals.groupby(['Question Item Id', 'Question Response', 'Undergrad Grad']).sum()
ugtotal.head(5)
```

```python
gender = data[['Question Stem Id', 'Question Item Id', 'Question Stem', 'Question Item', 'Question Response', 'Count', 'Demographic Value', 'Demographic Category']].drop_duplicates(ignore_index = True)
gender = gender.loc[gender['Demographic Category'] == 'Person Gender Desc']
gender = gender[gender['Question Item Id'].str.contains(question_stem, case=False)].reset_index(drop = True)
```

```python
# change values for all respondents option 
# Transgender Woman/Trans Woman + Woman --> Woman 
# Transgender Man/Trans Man + Man --> Man 
# Nonbinary + Different Identity --> Nonbinary 
```

### FUNCTION

```python
demos = data['Demographic Category'].unique()
qid = data['Question Item Id'].unique()

rand_qid = random.choice(qid)
rand_demos = random.choice(demos)

def demospotcheck(question_stem, demo_cat): 
    demototals = data.copy()
    demototals = demototals.loc[data['Demographic Category'] == demo_cat].drop_duplicates(ignore_index = True)
    demototals = demototals[demototals['Question Item Id'].str.contains(question_stem, case=False)]
    democounts = demototals.groupby(['Question Item Id', 'Demographic Value', 'Question Response']).agg({'Count': ['sum']})
    demoitemtotal = demototals[['Question Item Id', 'Demographic Value', 'Demographic Value Total']].drop_duplicates(ignore_index = True)
    demoall = pd.merge(democounts.reset_index(), demoitemtotal, on=['Question Item Id', 'Demographic Value'], how='inner')
    demoall.rename(columns={('Question Item Id', ''): "Question Item Id", ('Demographic Value', ''): "Demographic Value", ('Question Response', ''): "Question Response", ('Count', 'sum'): "Count"}, inplace=True)
    demoall['Percentage (%)'] = ((demoall['Count'] / demoall['Demographic Value Total']) * 100).round(2)
    demoall = demoall.loc[:,~demoall.columns.duplicated()]
    print('Question Item Id:', question_stem)
    print('Demographic Category:', demo_cat) 
    return demoall
```
